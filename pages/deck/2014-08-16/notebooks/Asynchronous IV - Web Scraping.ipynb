```python
import asyncio
import contextlib
from pyquery import PyQuery as pq
import requests
from urllib.parse import urlsplit
```

```python
URL = 'http://bugra.github.io' # Shameless promotion at its best
```

```python
r = requests.get(URL)
```

```python
print(r.ok, r.status_code) # Great news
```

```python
raw_html = r.text
```

```python
raw_html
```

```python
pp = pq(raw_html)
```

```python
posts = pp('ul')[2] ## 2 => Only Blog Posts 
```

```python
links = []
for post in posts:
    title = pq(post).find('a').text()[:-10]
    # Date is in the following Format YYYY.MM.DD , 10 characters
    date = pq(post).find('a').text()[-10:]
    link = pq(post).find('a').attr.href
    tup = (title, link, date)
    links.append(tup)
    print(' '.join(tup) + '\n')
```

```python
# Sort by date (corresponds to reverse as well as posts are already sorted by date) links = links[::-1]
links.sort(key=lambda k:k[2]) # inplace
```

```python
@asyncio.coroutine
def get_body(url):
    url = urlsplit(url)
    path = '/' * (not url.path) + url.path + '?' * bool(url.query) + url.query # Booleans are sub-class of integers
    reader, writer = yield from asyncio.open_connection(
        host=url.hostname,
        port=url.port or (443 if url.scheme == 'https' else 80),
        ssl=(url.scheme == 'https'))
    with contextlib.closing(writer):
        writer.write(b'GET ' + path.encode('ascii') + b' HTTP/1.1\r\n'
                     b'Host: ' + url.netloc.encode('ascii') + b'\r\n'
                     b'Connection: close\r\n\r\n')
        while True:
            line = yield from reader.readline()
            line = line.rstrip(b'\n\r')
            print(line.decode('latin-1'))
            if not line:
                break
        body = yield from reader.read()
    return body

#loop = asyncio.get_event_loop()
#body = loop.run_until_complete(get_body(URL))
#print(body)
#loop.close()

page_contents = []
@asyncio.coroutine
def inform(link_blob):
    title, url, date = link_blob
    page = yield from get_body(url) 

    ## Limit the resources with Semaphore
    """
    sem = asyncio.Semaphore(10) # Permit only 10 connections at a time
    with (yield from sem):
        page = yield from get_body(url)
    """
    
    page_contents.append((title, url, date, page))
    print('Date: {date}: titled {title} post is retrieved'.format(title=title, date=date))
```

```python
new_loop = asyncio.get_event_loop()
f = asyncio.wait([inform(link_blob) for link_blob in links])
new_loop.run_until_complete(f)
```
